```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib_venn import venn2, venn3
```

```python
df = pd.read_csv("important_genes.csv")
xg = set(df['XGBoost Genes'].dropna().tolist())
ab = set(df['AdaBoost Genes'].dropna().tolist())
lass = set(df['Lasso Convergence'].dropna().tolist())
```

```python
venn2([xg, ab], ('XGBoost', 'AdaBoost'))
plt.show()
```

```python
venn2([lass, ab], ('Lasso Convergence', 'AdaBoost'))
plt.show()
```

```python
venn2([lass, xg], ('Lasso Convergence', 'XGBoost'))
plt.show()
```

```python
venn3([xg, ab, lass], ('XGBoost', 'AdaBoost', 'Lasso Convergence'))
plt.show()
```
